```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Release_Year'] = df['Release_Date'].dt.year

# Filter data for years between 2008 and 2021
df_filtered = df[(df['Release_Year'] >= 2008) & (df['Release_Year'] <= 2021)]

# Group by year and category and count the occurrences
category_counts = df_filtered.groupby(['Release_Year', 'Category']).size().unstack(fill_value=0)

display(category_counts)
```

```python
# Split countries by comma and explode into separate rows
df['Country'] = df['Country'].str.split(', ')
df = df.explode('Country')

# Convert 'Release_Date' to datetime and extract 'Release_Year' after exploding
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Release_Year'] = df['Release_Date'].dt.year

# Group by Country and Release_Year
country_year = df.groupby(['Country', 'Release_Year']).size().reset_index(name='Count')

# Total content count per country
top_countries = country_year.groupby('Country')['Count'].sum().sort_values(ascending=False).head(6).index.tolist()

# Filter only top countries
filtered = country_year[country_year['Country'].isin(top_countries)]

display(filtered)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered, x='Release_Year', y='Count', hue='Country', marker='o')

plt.title('Top Country Contributions to Netflix Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

# Drop NaN values in 'Director' column
df_directors = df.dropna(subset=['Director'])

# Split multiple directors and flatten list
all_directors = df_directors['Director'].str.split(', ').explode()

# Count occurrences
director_counts = all_directors.value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=director_counts.values, y=director_counts.index, palette="Blues_d")
plt.title("Top 10 Most Featured Directors on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.tight_layout()
plt.show()
```

```python
# Filter out missing durations
df = df.dropna(subset=['Duration', 'Category'])

# Separate Movies and TV Shows
movies = df[df['Category'] == 'Movie'].copy()
tv_shows = df[df['Category'] == 'TV Show'].copy()
# Extract minutes for movies
movies['Minutes'] = movies['Duration'].str.extract('(\d+)').astype(float)

# Extract seasons for TV shows
tv_shows['Seasons'] = tv_shows['Duration'].str.extract('(\d+)').astype(float)
# Set plot style
sns.set(style="whitegrid")

# Create subplots
plt.figure(figsize=(14, 6))

# Histogram for Movie Duration
plt.subplot(1, 2, 1)
sns.histplot(movies['Minutes'], bins=30, kde=True, color='blue')
plt.title("Distribution of Movie Durations (Minutes)")
plt.xlabel("Minutes")
plt.ylabel("Number of Movies")

# Histogram for TV Show Seasons
plt.subplot(1, 2, 2)
sns.histplot(tv_shows['Seasons'], bins=10, kde=False, color='green')
plt.title("Distribution of TV Show Length (Seasons)")
plt.xlabel("Number of Seasons")
plt.ylabel("Number of TV Shows")

plt.tight_layout()
plt.show()
```
